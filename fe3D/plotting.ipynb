```python
import numpy as np
#from mpl_toolkits.mplot3d import Axes3D
from pylab import *
```

# Upload Data

```python
# Retreive L2 error data 

start = 2
end = 22
step = 1

# Normal

L2_err_dir = np.loadtxt('Data_err/L2error_degstart%(deg_start)d_degend%(deg_end)d_dir.txt'
    % {"deg_start":start,"deg_end":end})
L2_err_iter_prec = np.loadtxt('Data_err/L2error_degstart%(deg_start)d_degend%(deg_end)d_prec.txt'
    % {"deg_start":start,"deg_end":end})
L2_err_iter_noprec = np.loadtxt('Data_err/L2error_degstart%(deg_start)d_degend%(deg_end)d_noprec.txt'
    % {"deg_start":start,"deg_end":end})

# Matrix Free

start_mf = 2
end_mf = 35

L2_err_iter_prec_mf = np.loadtxt('Data_err_matfree/L2error_degstart%(deg_start)d_degend%(deg_end)d_prec.txt'
    % {"deg_start":start_mf,"deg_end":end_mf})
L2_err_iter_noprec_mf = np.loadtxt('Data_err_matfree/L2error_degstart%(deg_start)d_degend%(deg_end)d_noprec.txt'
    % {"deg_start":start_mf,"deg_end":end_mf})

```

```python
# Retreive timing data 

#Normal

t_dir = np.loadtxt('Data_time/Time_degstart%(deg_start)d_degend%(deg_end)d_dir.txt'
    % {"deg_start":start,"deg_end":end})
t_iter_prec = np.loadtxt('Data_time/Time_degstart%(deg_start)d_degend%(deg_end)d_prec.txt'
    % {"deg_start":start,"deg_end":end})
t_iter_noprec = np.loadtxt('Data_time/Time_degstart%(deg_start)d_degend%(deg_end)d_noprec.txt'
    % {"deg_start":start,"deg_end":end})

# Matrix Free

t_iter_prec_mf = np.loadtxt('Data_time_matfree/Time_degstart%(deg_start)d_degend%(deg_end)d_prec.txt'
    % {"deg_start":start_mf,"deg_end":end_mf})
t_iter_noprec_mf = np.loadtxt('Data_time_matfree/Time_degstart%(deg_start)d_degend%(deg_end)d_noprec.txt'
    % {"deg_start":start_mf,"deg_end":end_mf})
```

# L2 Error Plotting

Note: the tollerance of the iteractive method was set to 10^(-10)

```python
plt.figure()
plt.title('L2 error')
plt.semilogy(range(start, end, step), L2_err_dir, 'blue', label='Dir')
plt.semilogy(range(start, end, step), L2_err_iter_prec, 'green', label='Prec')
plt.semilogy(range(start, end, step), L2_err_iter_noprec, 'red', label='No Prec')
plt.xlabel('degree')
plt.ylabel('L2 error')
plt.legend(loc='lower left')
rcParams['figure.figsize'] = (10, 6)
plt.show()
    
```

# Timing Plot

Note: the tollerance of the iteractive method was set to 10^(-10)

```python
plt.figure()  
plt.title('Timing different methods')
plt.plot(range(start, end, step), t_dir, 'b', label='Dir')
plt.plot(range(start, end, step), t_iter_prec, 'g', label='Prec')
plt.plot(range(start, end, step), t_iter_noprec, 'r', label='No Prec')
plt.plot(range(start_mf, end_mf, step), t_iter_noprec_mf, 'y', linestyle='--', label='No Prec MFree')
plt.plot(range(start_mf, end_mf, step), t_iter_prec_mf, 'm', linestyle='--', label='Prec MFree')
plt.xlabel('degree')
plt.ylabel('Time')
plt.legend(loc = 'upper left')
rcParams['figure.figsize'] = (10, 6)
#plt.savefig('fig1.png', dpi = 300)
plt.show()
```
